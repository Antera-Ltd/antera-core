```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-muted')
ANTERA_COLORS = ['#FA520F', '#09090B', '#71717A']

```


<div style="padding: 10px 0; border-bottom: 2px solid #E4E4E7; margin-top: 40px; margin-bottom: 20px;">
    <h2 style="margin: 0; color: #09090B; font-size: 24px; display: flex; align-items: center;">
        <span style="margin-right: 15px;">🌍</span> II. Global Node Health Monitoring
    </h2>
</div>



```python
def simulate_telemetry(n_nodes=12):
    nodes = [f'Node-{i:02d}' for i in range(n_nodes)]
    data = []
    for node in nodes:
        for _ in range(50):
            data.append({
                'node': node,
                'latency': np.random.lognormal(2, 0.5) + (20 if node == 'Node-04' else 0),
                'cpu_util': np.random.uniform(40, 95),
                'uptime': np.random.choice([1, 0], p=[0.99, 0.01])
            })
    return pd.DataFrame(data)

df_tel = simulate_telemetry()
df_tel.groupby('node').mean()
```


<div style="padding: 10px 0; border-bottom: 2px solid #E4E4E7; margin-top: 40px; margin-bottom: 20px;">
    <h2 style="margin: 0; color: #09090B; font-size: 24px; display: flex; align-items: center;">
        <span style="margin-right: 15px;">📊</span> III. Latency Distribution & Outlier Detection
    </h2>
</div>



```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_tel, x='node', y='latency', palette='Oranges')
plt.axhline(df_tel['latency'].median(), color='#09090B', linestyle='--', label='Global Median')
plt.title('NODE LATENCY DISTRIBUTION (SLA COMPLIANCE CHECK)', loc='left', fontweight='bold')
plt.ylabel('Latency (ms)')
plt.legend()
plt.show()
```
